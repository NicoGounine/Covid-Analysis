```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as plt
import datetime as dt

diagnostico = pd.read_csv("../input/covid19-daily-case-hospitalization-spain-dataset/casos_diag_ccaadecl.csv")
caso =  pd.read_csv("../input/covid19-daily-case-hospitalization-spain-dataset/casos_hosp_uci_def_sexo_edad_provres.csv")
tecnica_CCAA = pd.read_csv("../input/covid19-daily-case-hospitalization-spain-dataset/casos_tecnica_ccaa.csv")
tecnica_provincia = pd.read_csv("../input/covid19-daily-case-hospitalization-spain-dataset/casos_tecnica_provincia.csv")
```

```python
tecnica_CCAA['fecha'] = pd.to_datetime(tecnica_CCAA.fecha)
plt.figure(figsize=(20, 5))
plt.plot(tecnica_CCAA["fecha"].loc[tecnica_CCAA["ccaa_iso"] == "CT"], tecnica_CCAA["num_casos"].loc[tecnica_CCAA["ccaa_iso"] == "CT"])
plt.title('Covid Cases in Catalonia')
plt.xlabel('Date')
plt.ylabel('Cases')
plt.show()

tecnica_CCAA['fecha'] = pd.to_datetime(tecnica_CCAA.fecha)
plt.figure(figsize=(20, 5))
plt.plot(tecnica_CCAA["fecha"].loc[tecnica_CCAA["ccaa_iso"] == "MD"], tecnica_CCAA["num_casos"].loc[tecnica_CCAA["ccaa_iso"] == "MD"])
plt.title('Covid Cases in Madrid')
plt.xlabel('Date')
plt.ylabel('Cases')
plt.show()
```
